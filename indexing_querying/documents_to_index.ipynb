```python
def convert_rating_to_text_atmosphere(rating):
    if rating==5.0:
        return 'Excellent Atmosphere'
    elif rating==4.0:
        return 'Good Atmosphere'
    elif rating==3.0:
        return 'Average Atmosphere'
    elif rating==2.0:
        return 'Poor Atmosphere'
    elif rating==1.0:
        return 'Terrible Atmosphere'
    else:
        return 'NaN'

def convert_rating_to_text_food(rating):   
    if rating==5.0:
        return 'Excellent Food'
    elif rating==4.0:
        return 'Good Food'
    elif rating==3.0:
        return 'Average Food'
    elif rating==2.0:
        return 'Poor Food'
    elif rating==1.0:
        return 'Terrible Food'
    else:
        return 'NaN'

def convert_rating_to_text_service(rating):
    if rating==5.0:
        return 'Excellent Service'
    elif rating==4.0:
        return 'Good Service'
    elif rating==3.0:
        return 'Average Service'
    elif rating==2.0:
        return 'Poor Service'
    elif rating==1.0:
        return 'Terrible Service'
    else:
        return 'NaN'
```

```python
all_merged['reviewDetailedRatingAtmosphere'] = all_merged.apply(lambda row: convert_rating_to_text_atmosphere(row['reviewDetailedRatingAtmosphere']), axis=1)
```

```python
all_merged['reviewDetailedRatingFood'] = all_merged.apply(lambda row: convert_rating_to_text_food(row['reviewDetailedRatingFood']), axis=1)
```

```python
all_merged['reviewDetailedRatingService'] = all_merged.apply(lambda row: convert_rating_to_text_service(row['reviewDetailedRatingService']), axis=1)
```

Convert locationLat and locationLong into a single LatLong variable for compatability with Solr spatial search

```python
def convert_to_latlong(d):
    lat = d['locationLat']
    long = d['locationLong']
    latlong = str(lat)+","+str(long)
    return latlong
```

```python
all_merged['LatLong'] = all_merged.apply(lambda row: convert_to_latlong(row), axis=1)
```

```python
all_merged = all_merged.drop(columns=['locationLat', 'locationLong'])
all_merged
```

To help in indexing later, we create another variable called qualityScore

The quality score is in the range [0,1] and is assigned to each document

It is comprised of two parts: one reviewer-dependent and another restaurant-dependent as shown in the formula below:

(raw) qualityScore = 0.4*(review_stars + likesCount + reviewerNumberOfReviews + isLocalGuide(1:True, 0:False)) + 0.6*(totalScore + reviewsCount)

We arbitrarily give the restaurant-dependent component a higher weighting of 0.6 as personal opinions on the restaurant might vary

We normalize all qualityScores by dividing using the maximum possible qualityScore in the entire collection

qualityScore = (0.4*(review_stars + likesCount + reviewerNumberOfReviews + isLocalGuide(1:True, 0:False)) + 0.6*(totalScore + reviewsCount))/max(qualityScore)

```python
def g(d):
    reviewer_dependent = d['review_stars'] + d['likesCount'] + d['reviewerNumberOfReviews']
    if d['isLocalGuide']:
        islocalguide = 1
    else:
        islocalguide = 0
    reviewer_dependent += islocalguide
    restaurant_dependent = d['totalScore'] + d['reviewsCount']

    return reviewer_dependent*0.4 + restaurant_dependent*0.6

def norm_g(d, maxg):
    return d['raw_qualityScore']/maxg
```

```python
all_merged['raw_qualityScore'] = all_merged.apply(lambda row: g(row), axis=1)
```

```python
all_merged
```

```python
maxg = all_merged.loc[all_merged['raw_qualityScore'].idxmax()]['raw_qualityScore']
all_merged['qualityScore'] = all_merged.apply(lambda row: norm_g(row, maxg), axis=1)
```

```python
all_merged = all_merged.drop(columns=['raw_qualityScore'])
all_merged
```

Finally, create the XML file to be posted into the collection used for Solr. Note that this XML file effectively contains ALL documents to be indexed

```python
from lxml import etree as et

root = et.Element("add")
for i in range(len(all_merged)):
    child = et.SubElement(root, "doc")
    field = et.SubElement(child, "field", name="id")
    field.text = str(i)
    for col in all_merged.columns:
        field = et.SubElement(child, "field", name=f"{col}")
        field.text = str(all_merged.iloc[i][col])

xml_data = et.tostring(root, pretty_print=True, xml_declaration=True)
with open('docs_to_post.xml', 'w') as f:
    f.write(xml_data.decode('utf-8'))  
    f.close() 
```
