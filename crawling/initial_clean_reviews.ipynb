```python
import pandas as pd
from langdetect import detect
```

```python
reviews = pd.read_csv("apify_all_reviews.csv")
reviews
```

```python
# drop reviews that do not contain any test or did not rate any stars
reviews.drop(reviews[(reviews.text.isnull()) | (reviews.stars.isnull())].index, inplace=True)
reviews
```

```python
def check_en(row):
    '''
    check if the review text is in English or not
    '''
    try:
        lang = detect(row['text'])
    except:
        lang = 'NA'
    return lang 
reviews["lang"] = reviews.apply(check_en, axis=1)
reviews
```

```python
# check how many languages are there for the reviews 
reviews.lang.unique()
```

```python
not_en = reviews.loc[reviews.lang != 'en']
not_en
```

```python
# iterate thru the reviews flagged as not English and confirm not English
idxs = []
for i in not_en.index:
    idxs.append(i)
```

```python
i = idxs[0]
print(f"index: {i}")
print(not_en['text'][i])
```

```python
id_to_drop = [854, 871, 872, 1105, 1138, 1139, 1220, 1911, 1928, 1999, 2356, 2375, 2430, 3494, 3942, 5379, 5380, 7506, 9256, 9511, 9512, 10691]
print(f"number of reviews to drop: {len(id_to_drop)}")
```

```python
reviews.drop(id_to_drop, axis=0, inplace=True)
reviews
```

```python
reviews['num_words'] = reviews.text.str.split().str.len()
print(f"total number of words for all reviews: {reviews['num_words'].sum(axis=0)}")
```

```python
print(f"total number of records/reviews: {reviews.shape[0]}")
```

```python
reviews.to_csv("cleaned_reviews.csv")
```
